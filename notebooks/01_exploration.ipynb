```python
# Imports
from pathlib import Path
import matplotlib.pyplot as plt
import random
from PIL import Image

DATA_DIR = Path("../data/raw")

# Get all image file paths from each class subfolder
image_paths = list(DATA_DIR.glob("*/*.png"))

# Count total number of images
print(f"Total images: {len(image_paths)}")

# Create a dictionary to count images per class
from collections import defaultdict
class_counts = defaultdict(int)

for path in image_paths:
    label = path.parent.name  # folder name = class label
    class_counts[label] += 1

# Show class distribution
for label, count in class_counts.items():
    print(f"{label}: {count} images")

# Summary plot of class counts
plt.figure(figsize=(6, 4))
plt.bar(class_counts.keys(), class_counts.values(), color='skyblue')
plt.title("Image Count per Class")
plt.xlabel("Class")
plt.ylabel("Number of Images")
plt.grid(True, axis='y', linestyle='--', alpha=0.6)
plt.show()

# Check a few image sizes
image_sizes = [Image.open(path).size for path in random.sample(image_paths, 10)]
print("Sample image sizes:")
for i, size in enumerate(image_sizes):
    print(f"  Image {i+1}: {size}")

# Display a few sample images from each class
def show_samples_per_class(image_paths, classes, n_samples=3):
    plt.figure(figsize=(n_samples * 3, len(classes) * 3))

    for row, cls in enumerate(classes):
        class_images = list((DATA_DIR / cls).glob("*.png"))
        selected = random.sample(class_images, n_samples)

        for col, img_path in enumerate(selected):
            img = Image.open(img_path)
            plt.subplot(len(classes), n_samples, row * n_samples + col + 1)
            plt.imshow(img)
            plt.title(f"{cls}")
            plt.axis("off")

    plt.tight_layout()
    plt.show()

# Run the function to visualize examples
class_labels = sorted(class_counts.keys())
show_samples_per_class(image_paths, class_labels)
```
