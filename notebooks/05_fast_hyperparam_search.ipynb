```python
# ‚îÄ‚îÄ‚îÄ Add Project Root to Python Path ‚îÄ‚îÄ‚îÄ
import sys, os
sys.path.insert(0, os.path.abspath(".."))

# ‚îÄ‚îÄ‚îÄ Imports ‚îÄ‚îÄ‚îÄ
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, callbacks
import keras_tuner as kt
import time

from src.models import build_mediumnet
```

```python
# ‚îÄ‚îÄ‚îÄ Define Model Builder for Learning Rate Tuning ‚îÄ‚îÄ‚îÄ
def model_builder(hp):
    # Search space: learning rate between 1e-4 and 1e-2 (log scale)
    hp_lr = hp.Float("learning_rate", min_value=1e-4, max_value=1e-2, sampling="log")

    model = build_mediumnet(input_shape=(150, 150, 3))
    model.compile(
        optimizer=tf.keras.optimizers.Adam(learning_rate=hp_lr),
        loss='sparse_categorical_crossentropy',
        metrics=['accuracy']
    )
    return model
```

```python
# ‚îÄ‚îÄ‚îÄ Fast Hyperparameter Search (Quick Mode: 4 trials √ó 3 epochs) ‚îÄ‚îÄ‚îÄ
tuner = kt.RandomSearch(
    model_builder,
    objective="val_accuracy",
    max_trials=4,
    executions_per_trial=1,
    directory="kt_dir",
    project_name="rps_mediumnet_fast"
)

# Show the search space
tuner.search_space_summary()

# Run the search on a small budget for speed
tuner.search(train_ds, validation_data=val_ds, epochs=3)

# Display results
tuner.results_summary()

# Extract best learning rate
best_hp = tuner.get_best_hyperparameters(1)[0]
best_lr = best_hp.get("learning_rate")
print("üìå Best learning rate found:", best_lr)
```

```python
# ‚îÄ‚îÄ‚îÄ Final Training Run with Best LR (MediumNet-fast) ‚îÄ‚îÄ‚îÄ
model = build_mediumnet(input_shape=(150, 150, 3))

model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=best_lr),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

early_stop = callbacks.EarlyStopping(patience=2, restore_best_weights=True)

print("üöÄ Training MediumNet-fast with tuned learning rate...")
start = time.time()

history = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=10,
    callbacks=[early_stop],
    verbose=2
)

train_time = time.time() - start
print(f"‚è±Ô∏è  MediumNet-fast training time: {train_time:.1f} sec")

# Final evaluation
test_loss, test_acc = model.evaluate(test_ds, verbose=0)
print(f"üéØ MediumNet-fast test accuracy: {test_acc:.4f}, loss: {test_loss:.4f}")
```
